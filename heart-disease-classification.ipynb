# Predicting heart disease using machine learning
This notebook explores how machine learning can be applied to clinical data in order to predict whether or not a patient has heart disease. The goal is to use various Python-based data science tools to analyze medical attributes and evaluate classification models.

## 1: Problem definition:
The goal is to identify patterns in medical attributes that can help predict whether a patient has heart disease.  

**In short:**  
> Given clinical parameters about a patient, can we predict whether or not they have heart disease?


## 2- Data:
The dataset comes from the [UCI Machine Learning Repository â€“ Heart Disease](https://archive.ics.uci.edu/dataset/45/heart+disease).  

It contains various clinical attributes (such as age, sex, cholesterol, chest pain type, etc.) that are commonly used in diagnosing heart disease.


## 3- Features 

Below is the data dictionary describing each feature in the dataset:

* age: age in years
* sex: sex (1 = male; 0 = female)
* cp: chest pain type
   -- Value 1: typical angina
   -- Value 2: atypical angina
   -- Value 3: non-anginal pain
   -- Value 4: asymptomatic
* trestbps: resting blood pressure (in mm Hg on admission to the hospital)
* chol: serum cholestoral in mg/dl
* fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* restecg: resting electrocardiographic results
   -- Value 0: normal
   -- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
   -- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
* thalach: maximum heart rate achieved
* exang: exercise induced angina (1 = yes; 0 = no)
* oldpeak = ST depression induced by exercise relative to rest
* slope: the slope of the peak exercise ST segment
   -- Value 1: upsloping
   -- Value 2: flat
   -- Value 3: downsloping
* ca: number of major vessels (0-3) colored by fluoroscopy
* thal: 3 = normal; 6 = fixed defect; 7 = reversible defect
* num: diagnosis of heart disease (angiographic disease status)
   -- Value 0: < 50% diameter narrowing
   -- Value 1: > 50% diameter narrowing

## Machine Learning Model

```python
np.random.seed(42)
X=df.drop("target",axis=1)
y=df["target"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

### Data Preprocessing
To improve model performance, we scale the features so that values like age, cholesterol, and blood pressure are on a similar range.

```python
from sklearn.preprocessing import StandardScaler
np.random.seed(42)
scaler= StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

## 4. Modeling
We will train and evaluate several classification algorithms:
- Logistic Regression
- Random Forest
- Support Vector Machine (SVM)
- K-Nearest Neighbors (KNN)


```python

from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

models = {"RandomForest":RandomForestClassifier(),"LogisticRegression":LogisticRegression()
          ,"SVC":SVC(),"KNN":KNeighborsClassifier()}
def my_models (model,X_train,X_test,y_train,y_test):
    model_scores={}
    for name,model in models.items():
        model.fit(X_train,y_train)
        model_scores[name] = model.score(X_test,y_test)
    models_scores= pd.DataFrame(model_scores,index=["score"]).T
    print(models_scores)

my_models(models,X_train_scaled,X_test_scaled,y_train,y_test)
```

```python
from sklearn.metrics import classification_report

def my_models(models, X_train, X_test, y_train, y_test):
    model_scores = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = model.score(X_test, y_test)
        report = classification_report(y_test, y_pred, output_dict=True)
        model_scores[name] = {
            "Accuracy": acc,
            "Precision": report["weighted avg"]["precision"],
            "Recall": report["weighted avg"]["recall"],
            "F1-score": report["weighted avg"]["f1-score"]
        }
    return pd.DataFrame(model_scores).T

my_models(models, X_train_scaled, X_test_scaled, y_train, y_test)
```

## 5. Conclusion
**K-Nearest Neighbors (KNN)** achieved the highest accuracy (~90%), outperforming Logistic Regression, Random Forest, and SVM.  
This suggests that KNN may capture local patterns in clinical attributes better than the other models.
